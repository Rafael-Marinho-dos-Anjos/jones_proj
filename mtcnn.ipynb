```python
from facenet_pytorch import MTCNN
import cv2
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
from tqdm.notebook import tqdm
import torch
```

```python
# v_cap = cv2.VideoCapture(0)
# success, frame = v_cap.read()
# im_name = r"sample_images\download.jpeg"
im_name = r"sample_images\28287431-grupo-do-multietnico-o-negocio-pessoas-dentro-a-escritorio-ai-gerado-foto.jpg"

frame = cv2.imread(im_name)
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 8))
plt.imshow(frame)
plt.axis('off')
```

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
print(device)

# Create face detector
mtcnn = MTCNN(select_largest=False, device=device, keep_all=True)

# Load a single image and display
frame = Image.fromarray(frame)

# Detect face
faces = mtcnn(frame)
faces.shape
```

```python

# Visualize
fig, ax = plt.subplots(3, 5, figsize=(15, 9))

ax[0][0].set_ylabel("Image")
ax[1][0].set_ylabel("GT")
ax[2][0].set_ylabel("Mask")

for j in range(5):
    ax[0][j].imshow(faces[j].permute(1, 2, 0).numpy(), cmap="gray", interpolation="None")
    ax[0][j].set_xticks([])
    ax[0][j].set_yticks([])

    ax[1][j].imshow(faces[j+5].permute(1, 2, 0).numpy(), cmap="gray", interpolation="None")
    ax[1][j].set_xticks([])
    ax[1][j].set_yticks([])

    ax[2][j].imshow(faces[j+10].permute(1, 2, 0).numpy(), cmap="gray", interpolation="None")
    ax[2][j].set_xticks([])
    ax[2][j].set_yticks([])
```
